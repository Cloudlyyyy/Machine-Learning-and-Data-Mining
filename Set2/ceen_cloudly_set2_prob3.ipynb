```python
import numpy as np
from matplotlib import pyplot as plt
from numpy import genfromtxt
from sklearn import linear_model
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
%matplotlib inline
```

## Load data

```python
import urllib.request
urllib.request.urlretrieve('https://raw.githubusercontent.com/emiletimothy/Caltech-CS155-2023/main/set2/data/problem3data.txt', 'problem3data.txt')

train_data = np.loadtxt('problem3data.txt')
X_train = train_data[:, :9]
y_train = train_data[:, 9]
```

```python
alpha = np.linspace(0.01, 3, 30)

l_coef_list =[]
for i in range(len(alpha)):
  lasso = linear_model.Lasso(alpha = alpha[i]).fit(X_train, y_train)
  l_coef_list.append(lasso.coef_)

r_coef_list = []
for i in range(int(1e4)):
  ridge = linear_model.Ridge(alpha = i).fit(X_train, y_train)
  r_coef_list.append(ridge.coef_)

plt.plot(alpha, l_coef_list)
plt.xlim(0,4)
plt.legend(["w1", "w2", "w3", "w4", "w5", "w6", "w7", "w8", "w9"])
plt.xlabel("alpha")
plt.ylabel("weights")
plt.show()

plt.plot(range(int(1e4)), r_coef_list)
plt.xlim(0,12000)
plt.legend(["w1", "w2", "w3", "w4", "w5", "w6", "w7", "w8", "w9"])
plt.xlabel("alpha")
plt.ylabel("weights")
plt.show()
```
